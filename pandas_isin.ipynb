```python
#fas-정재환
import pandas as pd
import numpy as np
import csv
%matplotlib inline
```

```python
# weight는 체중이 65, 70, 75 인 데이터만 존재한다
weight = pd.read_csv('weight.csv', encoding='cp949')
weight
```

```python
# 반면에 df는 체중이 35부터 135까지 많은 경우의 체중을 가진 데이터가 존재한다
df = pd.read_csv('NHIS_OPEN_GJ_2017_4.csv', encoding='cp949', nrows=200000)
df
```

```python
# 그럼 df에서 체중이 65, 70, 75 인 데이터만 추출해보자
# 가장 간단하고 효율적인 방법은 isin()을 활용하는 것이다

# weight의 체중 컬럼을 나열한 배열을 list로 변환하고 해당 리스트에 해당하는 체중 값을 가진 데이터만 뽑을 수 있다
ar = weight['체중']
ar = list(ar)
df = df[df['체중'].isin(ar)]
df.head(50)
```
